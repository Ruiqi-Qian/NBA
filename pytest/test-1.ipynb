```python
import types

# 运行一遍你的 data_cleaning.py，把里面的变量读到 namespace
namespace = types.SimpleNamespace()
with open('C:/adv/data_cleaning.py', encoding='utf-8') as f:
    code = f.read()
exec(code, namespace.__dict__)

def test_no_missing_values_in_key_dataframes():
    """Test that key DataFrames from data_cleaning.py have no missing values."""

    important_dfs = [
        'df_per_ts', 
        'df_salary', 
        'df_selectd', 
        'df_pos',
        'df_pos_2019',
        'df_pos_2020',
        'df_pos_2021',
        'df_pos_2022',
        'df_merged_data_2019',
        'df_merged_data_2020',
        'df_merged_data_2021',
        'df_merged_data_2022'
    ]

    for df_name in important_dfs:
        df = getattr(namespace, df_name)
        missing = df.isnull().sum().sum()
        print(f"{df_name}: missing values = {missing}")
        assert missing == 0, f"❌ {df_name} contains missing values!"
```
